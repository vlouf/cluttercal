```python
import numpy as np
```

```python
def get_text(rid, year):
    txt = f'''#!/bin/bash
#PBS -P kl02
#PBS -q normal
#PBS -l walltime=0:30:00
#PBS -l mem=128GB
#PBS -l wd
#PBS -l jobfs=1GB
#PBS -l ncpus=16
#PBS -lstorage=scratch/kl02+gdata/rq0

source activate radar

python national_archive.py -t 40 -r {rid:02} -s {year}0701 -e {year}1231
'''
    return txt
```

```python
data = np.genfromtxt('../../suncal/notebooks/data/rid_year.txt', dtype=int)
```

```python
for rid, year in data:
    if year != 2020:
        continue
    txt = get_text(rid, year)
    with open(f'../scripts/qrca.{rid:02}.{year}.pbs', 'w+') as fid:
        fid.write(txt)
```
